```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
import time
import re 
```

```python
from bs4 import BeautifulSoup

#페이지 자르기(pageIdx 기준)
url1 = "https://www.oliveyoung.co.kr/store/display/getMCategoryList.do?dispCatNo=100000100020006&fltDispCatNo=&prdSort=01&pageIdx="
url2 ="&rowsPerPage=48&searchTypeSort=btn_thumb&plusButtonFlag=N&isLoginCnt=0&aShowCnt=0&bShowCnt=0&cShowCnt=0&trackingCd=Cat100000100020006_MID"

code_list = []

for a in range(1,6):
    url = url1 + str(a) + url2
    res = requests.get(url)
    html=res.text
    soup=BeautifulSoup(html, 'html.parser')

    for i in range(8, 21):
        #8-13까지 변경하면서
        for j in range(1, 5):
            if j == 1:
                codes=soup.select("#Contents ul:nth-child(%s) li.flag" %i)
            else:
                codes=soup.select("#Contents ul:nth-child(%s) li:nth-child(%s)" %(i,j))
                
            for code in codes:
                num = codes[0].select_one('a[href]')
                lists = re.findall(r'\d{12}', num['href'])
            code_list.append(lists[0])
            #print(a,"페이지", i, "번줄", j, "번째", lists[0])
code_list
```

```python
from collections import OrderedDict
code_list=list(OrderedDict.fromkeys(code_list))
len(code_list)
```

```python
import pandas as pd
df = pd.DataFrame(columns=['name', 'brand', 'price', 'ingredient'])

url1 = "https://www.oliveyoung.co.kr/store/goods/getGoodsDetail.do?goodsNo="
url2 = "&dispCatNo=1000001000200060003&trackingCd=Cat100000100020006_MID"

for i in range(len(code_list)):
    driver = webdriver.Chrome('/Users/yhg31/chromedriver.exe')
    try:
        url = url1 + "A" +code_list[i] + url2
        driver.get(url)
    
    except:
        url = url1 + "B" +code_list[i] + url2
        driver.get(url)
    
    try:
        name=driver.find_element_by_xpath('''/html/body/div[2]/div[7]/div/div[2]/div[2]/div/p[2]''').text.strip()
        brand=driver.find_element_by_xpath('''/html/body/div[2]/div[7]/div/div[2]/div[2]/div/p[1]''').text.replace(",", " ")
        price=driver.find_element_by_xpath('''//*[@id="Contents"]/div[2]/div[2]/div/div[1]/span/strong''').text.strip() 
        ingredient=driver.find_element_by_xpath('''//*[@id="buyInfo"]''').click()
        try :
            ingredient=driver.find_element_by_xpath('''/html/body/div[2]/div[7]/div/div[7]/div[2]/dl[7]/dd''').text
        except:
            ingredient=driver.find_element_by_xpath('''/html/body/div[2]/div[7]/div/div[8]/div[2]/dl[7]/dd''').text
    except:
        ingredient=0
    df=df.append({'name' : name, 'brand' : brand, 'price' : price, 'ingredient' : ingredient} , ignore_index=True)
    print(i, name, brand, price, "\n\n",ingredient)
    driver.close()
```
